```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = []
for i in range(1000):
    df.append(random.uniform(0, 10))
df = pd.DataFrame(df)
df = df.rename(columns={0 : 'GRADE'})
print(df)
```

```python
x = [i for i in range(11)]

plt.hist(df['GRADE'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Гистограмма распределения')
plt.xlabel('Числа')
plt.ylabel('Частота')
plt.xticks(x)
plt.grid(axis='y')
print(df)
plt.show()
```

```python
normalDf = np.random.normal(loc=0, scale=1, size=1000)
normalDf = pd.DataFrame(df)
normalDf = df.rename(columns={0 : 'GRADE'})

min_val = normalDf.min()
max_val = normalDf.max()

normalizedDf = 10 * (normalDf - min_val) / (max_val - min_val)
print(normalizedDf)
```

```python
plt.hist(normalizedDf['GRADE'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Гистограмма распределения')
plt.xlabel('Числа')
plt.ylabel('Частота')
plt.xticks(x) 
plt.grid(axis='y')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(normalizedDf['GRADE'], color='blue', label='Плотность распределения', bw_adjust=0.5)
plt.title('Плотность распределения нормализованных данных')
plt.xlabel('Нормализованное значение')
plt.ylabel('Плотность')
plt.legend()
plt.grid()
plt.show()
```

```python
def decreasingDistribution(columnsCount):
    columnsCount += 1
    x = 1000
    fib1, fib2 = 1,1
    xFib = []
    randomList = []
    xListDsit = []

    for i in range(2, columnsCount + 3):
        fib1, fib2 = fib2, fib1 + fib2
        xFib.append(fib2)
    for i in range(1000):
        randomList.append(random.uniform(0, xFib[columnsCount -1]))
    for i in range(columnsCount):
        if 0 < i <= columnsCount - 1:
            xListDsit.append(len(list(filter(lambda x: xFib[i - 1] < x < xFib[i], randomList))))
    return(xListDsit)
```
